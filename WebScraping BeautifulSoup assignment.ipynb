Q-1) Write a python program to display all the header tags from wikipedia.org and make data frame.

```python
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

```python
soup=BeautifulSoup(page.content)
tags=[]
for i in soup.find_all(["h1", "h2", "h3", "h4","h5"]):
    tags.append(i.text)

df=pd.DataFrame({'Tags':tags})
df
```

Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and
make data frame-

i) Headline

ii) Time

iii) News Link

```python
page=requests.get("https://www.cnbc.com/world/?region=world")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
headline=[]
time=[]
link=[]
for i in soup.find_all('a', class_="LatestNews-headline"):
    headline.append(i.text)
    

for i in soup.find_all('time', class_="LatestNews-timestamp"):
    time.append(i.text)

for i in soup.find_all('div', class_="LatestNews-headlineWrapper"):
    links = i.find("a")["href"]
    link.append(links)
    
df=pd.DataFrame({"Headline": headline, "Time": time, "Links": link})

df.head(10)
```

Write a python program to scrape the details of most downloaded articles from AI in last 90
days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles
Scrape below mentioned details and make data frame-

i) Paper Title

ii) Authors

iii) Published Date

iv) Paper URL

```python
page=requests.get("https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles")
page
```

```python
soup=BeautifulSoup(page.content, "html.parser")
soup
```

```python
paper_title=[]
Authors=[]
Published_date=[]
link=[]

for i in soup.find_all('a', class_="sc-5smygv-0 fIXTHm"):
    paper_title.append(i.text)
    
for i in soup.find_all('span', class_="sc-1w3fpd7-0 dnCnAO"):
    Authors.append(i.text)
    
for i in soup.find_all('span', class_="sc-1thf9ly-2 dvggWt"):
    Published_date.append(i.text)
    
for i in soup.find_all('li', class_="sc-9zxyh7-1 sc-9zxyh7-2 kOEIEO hvoVxs"):
    links = i.find("a")["href"]
    link.append(links)
    
df=pd.DataFrame({"Paper_title": paper_title, "Authors": Authors, "Published_date": Published_date, "Paper_URL":link})

df.head(10)
```

 Write a python program to scrape mentioned details from dineout.co.in and make data frame-
    
i) Restaurant name

ii) Cuisine

iii) Location

iv) Ratings

v) Image URL

```python
page=requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
soup=BeautifulSoup(page.content)

name=[]
cuisine=[]
loc=[]
rate=[]
link=[]

for i in soup.find_all('div', class_="restnt-info cursor"):
    name.append(i.text)

for i in soup.find_all('span', class_="double-line-ellipsis"):
    cuisines=i.find("a")
    cuisine.append(cuisines.text)
    
for i in soup.find_all('div', class_="restnt-loc ellipsis"):
    loc.append(i.text)

for i in soup.find_all('div', class_="restnt-rating rating-4"):
    rate.append(i.text)

for i in soup.find_all('img', class_="no-img"):
    link.append(i['data-src'])


df=pd.DataFrame({"Restaurant name": name, "Cuisine":cuisine, "Location":loc, "Ratings":rate, "Image URL":link})
df
```
